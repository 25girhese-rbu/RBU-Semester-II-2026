## Solve $x-2y+3z=2, 2x-3z=3, x+y+z=0$

```python
I=matrix(3,3,[1,-2,3,2,0,-3,1,1,1])
show(I)
```

```python
J=matrix(3,1,[2,3,0])
show(J)
```

```python
C=I.augment(J) # To augment I matrix in J
show(C)
```

```python
show(C.echelon_form())
```

```python
rank(I) #for Rank
```

```python
rank(C)
```

```python
rank(I)==rank(C)
```
