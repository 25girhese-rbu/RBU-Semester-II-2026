Example 1: Write a recurrence relation for the number of ways to arrange n distinct objects in a row.\n
Ans: Recurrence relation is $Y_{n} = nY_{(n-1)}$

```python
from sympy import Function, rsolve
var('n')
y = Function('y')
f = y(n) - n*y(n-1)
b = rsolve(f, y(n), {y(1): 1})
b
```

Example 2:An elf has a staircase of n stairs to climb. Each step it takes can cover either one stair or two stairs. Find a recurrence relation for the number of different ways for the elf to climb the n-
stair staircase and also solve thar recurrence relation.
Ans: Recurrence relation is $𝑌_{(𝑛)}=𝑌_{(𝑛−1)}+ Y_{(n-2)}$

```python
from sympy import Function, rsolve
var('n')
y = Function('y')
f = y(n)-y(n-1) - y(n-2)
b = rsolve(f, y(n), {y(1): 1, y(2): 2})
b
```

Example 3:The Tower of Hanoi is a game consisting of n circular rings of varying size and three pegs on which the rings fit. The rules of puzzle allow disks to be moved on at a time from one peg to another as long as a disk is never placed on top of a smaller disk. The goal of puzzle is to have all the disks on second pegin order of size, with the largest on bottom.
Ans: Recurrence relation is $Y_{(n)}= 2Y_{(n-1)}+1$

```python
from sympy import Function, rsolve
var('n')
y = Function('y')
f = y(n)-2*y(n-1)-1
b = rsolve(f, y(n), {y(1): 1})
b
```

Example 4: Suppose we draw n straight lines on a piece of paper so that every pair of lines intersect (but no three lines intersect at a common point). Into how many regions do these n lines divide the plane?
Ans: Recurrence relation is $Y_{(n)} = Y_{(n-1)}+n$

```python
from sympy import Function, rsolve
var('n')
y = Function('y')
f = y(n)-y(n-1)-n
b = rsolve(f, y(n), {y(0): 1})
b
```
