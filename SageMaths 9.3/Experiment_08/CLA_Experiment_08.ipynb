### Exercise — Local Max & Min (Single Variable)
Q1.Find  local  max  and  min of $(x)=x4e−x$

```python
f(x) = x^4 * exp(-x)
plot(f, -1, 6, figsize=4)

df  = f.diff()
d2f = f.diff(2)
cp  = solve(df == 0, x)
show("Critical points:", cp)

for pt in cp:
    val = d2f.subs(pt)
    show("At", pt, "-> f''=", val, "-> Max" if val < 0 else "Min" if val > 0 else "Inconclusive")

show("f(0) =", f(0))
show("f(4) =", f(4))
```

Q2. Find speed $v$
v that minimizes power $P=Av3+BL2vP = Av^3 + \dfrac{BL^2}{v}
P=Av3+vBL2​$

```python
var('v A B L')
assume(A > 0, B > 0, L > 0, v > 0)
P(v) = A*v^3 + B*L^2/v
dP = P.diff(v)
v_min = solve(dP == 0, v)
show("Speed minimizing power:", v_min)
show("Minimum power:", P(v).subs(v_min[0]))
```

### Exercise — Maxima & Minima Without Constraints
Q1. $f(x,y)=x^4+y^4−4*x^2-4*y^2$

```python
var('x y')
f(x,y) = x^4 + y^4 - 4*x^2 - 4*y^2
fx = f.diff(x);  fy = f.diff(y)
R = f.diff(x,2); S = f.diff(x,y); T = f.diff(y,2)
H = R*T - S^2
cp = solve([fx==0, fy==0], x, y)
show("Critical points:", cp)
for pt in cp:
    h_val = H.subs(pt)
    r_val = R.subs(pt)
    show(pt, "-> H=", h_val, ", fxx=", r_val)
```

### Exercise — Lagrange Multipliers
Q1. Extremum of $f(x,y)=x^2+y^2$ subject to
$x+y=5$

```python
var('x y t')
f(x,y) = x^2 + y^2
g(x,y) = x + y - 5
L(x,y,t) = f(x,y) + t*g(x,y)
cp = solve([L.diff(x)==0, L.diff(y)==0, L.diff(t)==0], x, y, t)
show("Critical point:", cp)
show("Extremum value:", f.subs(cp[0]))
```

Q2. Extremum of $f(x,y)=xy$ subject to $x+y=20$

Q3. Extremum of $f(x,y)=x+2y$ subject to $x^2+y^2=6$
